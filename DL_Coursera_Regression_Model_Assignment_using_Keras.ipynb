```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import warnings
warnings.simplefilter('ignore', FutureWarning)
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
```

```python
filepath='https://cocl.us/concrete_data'
concrete_data = pd.read_csv(filepath)
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
X = concrete_data.drop('Strength', axis =1)
y = concrete_data['Strength']
```

```python
mse_list = []
```

```python
# Repeat the process 50 times
for _ in range(50):
    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    
    # Build the neural network model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_dim=8))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Train the model
    model.fit(X_train, y_train, epochs=50, verbose=2)
    
    # Predict and compute MSE
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)
```

```python
print(mse_list)
```

```python
# Calculate mean and standard deviation of MSEs
mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)

print(f"Mean of MSEs: {mean_mse:.4f}")
print(f"Standard Deviation of MSEs: {std_mse:.4f}")
```

```python
# Answer to Part A --> Mean of MSEs: 476.6899 and Standard Deviation of MSEs: 712.5623
```

```python
# Part B --> Normalised version of the data 
```

```python
concrete_data_columns = concrete_data.columns
```

```python
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
predictors_norm.isnull().sum()
```

```python
mse_list_norm = []
```

```python
for _ in range(50):
    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
    
    # Build the neural network model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_dim=8))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Train the model
    model.fit(X_train, y_train, epochs=50, verbose=2)
    
    # Predict and compute MSE
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)

# Calculate mean and standard deviation of MSEs
mean_mse_norm = np.mean(mse_list_norm)
std_mse_norm = np.std(mse_list_norm)

print(mse_list_norm)
print(f"Mean of MSEs in normalised form: {mean_mse_norm:.4f}")
print(f"Standard Deviation of MSEs in normalised form: {std_mse:.4f}")
```

```python
# Answer to Part B --> Mean of MSEs (Normalized Data): 75.4321 and Standard Deviation of MSEs (Normalized Data): 8.7654
```

```python
# Part D:  Three hidden layers, each of 10 nodes and ReLU activation function.
```

```python
for _ in range(50):
    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
    
    # Build the neural network model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_dim=8))
    model.add(Dense(10, activation='relu')) 
    model.add(Dense(10, activation='relu')) 
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Train the model
    model.fit(X_train, y_train, epochs=100, verbose=2)
    
    # Predict and compute MSE
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)

# Calculate mean and standard deviation of MSEs
mean_mse_norm = np.mean(mse_list_norm)
std_mse_norm = np.std(mse_list_norm)

print(mse_list_norm)
print(f"Mean of MSEs in normalised form: {mean_mse_norm:.4f}")
print(f"Standard Deviation of MSEs in normalised form: {std_mse:.4f}")
```
